```python
# Лабораторная работа №7
# Визуализация датасета

import networkx as nx
import matplotlib.pyplot as plt
import os
import matplotlib

    
G = nx.read_gexf('EuroSiS Generale Pays.gexf')
options = {
    'node_color': 'blue',
    'node_size': 60,
    'width': 0.5,
    'with_labels': True,
    'alpha': 0.6,
    'font_size': 3,
    'font_family': 'Arial'
}
       
plt.plot()
plt.title('kamada_kawai')
nx.draw_kamada_kawai(G, **options)

plt.tight_layout()
plt.show()
```

```python
plt.plot()
plt.title('circular')
nx.draw_circular(G, **options)

plt.tight_layout()
plt.show()
```

```python
plt.plot()
plt.title('spectral')
nx.draw_spectral(G, **options)
    
plt.tight_layout()
plt.show()
```

```python
plt.plot() 
plt.title('spring')
nx.draw_spring(G, **options)
    
plt.tight_layout()
plt.show()
```
