```python
# 2D

import pandas as pd
import requests
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt

url = "https://raw.githubusercontent.com/a-vodka/dv/master/lab/dataset/4/transfusion.data"
c = pd.read_csv(url, sep = ",", engine='python')

pca = PCA(n_components=5, whiten=True)
pca.fit(c)

x_pca = pca.transform(c)

for i in x_pca:
    plt.scatter(i[0], i[1])

plt.show()
```

```python
# 3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

pca = PCA(n_components=5, whiten=True)
pca.fit(c)

x_pca = pca.transform(c)

for i in x_pca:
    ax.scatter(i[0], i[1], i[2])

plt.show()
```
