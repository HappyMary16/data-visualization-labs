```python
# Вариант 1
# Построить столбиковую 2d диаграмму

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple

years = (1900, 1913, 1929, 1938, 1950, 1960, 1970, 1980, 1990, 2000)
years_n = len(years)

USA = (76.4, 97.6, 122.2, 130.5, 153, 176, 200.5, 227, 247, 277)
Germany = (47.5, 54.7, 58.4, 62.3, 67, 72, 77, 78.5, 79, 82)
France = (40.8, 41.8, 42, 42, 42, 46, 50.5, 54, 56.5, 59)
Japan = (44, 51.6, 63.2, 71.8, 83, 93, 104, 116.8, 123.5, 127)
USSR = (123, 158, 171.5, 186.5, 205.5, 226.5, 247, 258.5, 290, 290)

fig, ax = plt.subplots()
index = np.arange(0, years_n * 2.5, 2.5)
bar_width = 0.35
opacity = 0.4

rects1 = ax.bar(index, USA, bar_width,
                alpha=opacity, color='b',
                label='USA')

rects2 = ax.bar(index + bar_width, Germany, bar_width,
                alpha=opacity, color='r',
                label='Germany')

rects3 = ax.bar(index + bar_width * 2, France, bar_width,
                alpha=opacity, color='g',
                label='France')

rects3 = ax.bar(index + bar_width * 3, Japan, bar_width,
                alpha=opacity, color='y',
                label='Japan')

rects3 = ax.bar(index + bar_width * 4, USSR, bar_width,
                alpha=opacity, color='black',
                label='USSR')

ax.set_xlabel('Years')
ax.set_ylabel('Population')
ax.set_title('Population')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(years)
ax.legend()

plt.show()
```

```python
# Вариант 1
# Построить столбиковую 3d диаграмму

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

chart = plt.figure()
ax = chart.add_subplot(111, projection='3d')

ySpace = 0

def buildBar(z, x, c):
    global ySpace
    X = np.array([x, x])
    Y = np.array([ySpace, ySpace])
    Z = np.array([[z, 0]])
    ax.plot_wireframe(X, Y, Z, color=c,rstride=15, cstride=10)
    
    ySpace += 10

# Строим 3d диаграмму
for i in range(years_n):
    buildBar(USA[i], years[i], 'b')
    buildBar(Germany[i], years[i], 'r')
    buildBar(France[i], years[i], 'g')
    buildBar(Japan[i], years[i], 'y')
    buildBar(USSR[i], years[i], 'black')
    ySpace = 0

ax.set_xlabel('Years')
ax.set_ylabel('Countries')
ax.set_zlabel('Population')
ax.set_title('Population')

plt.show()
```
