```python
# Вариант 1
# График функции y

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-1, 0, 0.01)
y = np.cos(x - 1) * np.exp(6 * x)

plt.title("X, Y") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.plot(x, y) 
plt.show()
```

```python
# График функции z

x1 = np.arange(-1, 0, 0.01)
x2 = np.arange(0.01, 1, 0.01)

z1 = (1 + pow(x1, 2)) / (pow((1 + pow(x1, 4)), 1 / 2))
z2 = 2 * x2 + pow(np.sin(x2), 2) / (2 + x2)

plt.plot(x1, z1, 'g-')
plt.plot(x2, z2, 'g-')

plt.title("X, Z") 
plt.xlabel('x')
plt.ylabel('z')

plt.show()
```
