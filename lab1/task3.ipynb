```python
# Кардиоида
# Построить график в полярной системе координат

import numpy as np
import matplotlib.pyplot as plt


a = 1
fi = np.arange(0, 7, 0.01)
theta = 2 * (1 + np.cos(fi))

ax = plt.subplot(111, projection='polar')
ax.plot(fi, theta)
ax.set_rmax(5)
ax.set_rticks([0, 1, 2, 3, 4])  # less radial ticks
ax.grid(True)
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.set_title("A line plot on a polar axis")
plt.show()
```
