```python
# Вариант 1
# Построить график в полярной системе координат

import numpy as np
import matplotlib.pyplot as plt


f = 2
a = np.arange(0, 2, 0.01)
theta = 3 * a * np.cos(f) * np.sin(f) / (pow(np.cos(f), 3) + pow(np.sin(f), 3))

ax = plt.subplot(111, projection='polar')
ax.plot(theta, a)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # less radial ticks
ax.grid(True)
ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.set_title("A line plot on a polar axis")
plt.show()
```
